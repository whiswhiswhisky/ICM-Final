```python
import requests
```

```python
resp = requests.get("https://www.quora.com/What-does-your-dream-life-look-like")
html_str = resp.text
```

```python
from bs4 import BeautifulSoup
```

```python
document = BeautifulSoup(html_str,"html.parser")
type(document)
```

```python
para = document.find_all('p',attrs={"class":"ui_qtext_para u-ltr u-text-align--start"})
para 
```

```python
text = document.find_all('span',attrs={"class":"ui_qtext_rendered_qtext"})
text
```

```python
answers = [item.string for item in text] 
```
